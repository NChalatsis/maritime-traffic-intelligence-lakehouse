```python
from pyspark.sql import functions as F

synthetic_df = spark.table("bronze_ais_synthetic")
noaa_df = spark.table("bronze_ais_noaa")

print(f"Synthetic rows: {synthetic_df.count()}")
print(f"NOAA rows: {noaa_df.count()}")
```

```python
canonical_bronze_df = synthetic_df.unionByName(noaa_df, allowMissingColumns=True)

print(f"Canonical bronze rows: {canonical_bronze_df.count()}")
display(canonical_bronze_df.limit(10))
```

```python
canonical_bronze_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_ais_canonical")
```

```python
spark.table("bronze_ais_canonical").printSchema()
display(spark.table("bronze_ais_canonical").limit(10))
```
